```python
# ==========================================
# SHOPPING DATASET ANALYSIS PROJECT
# ==========================================

# STEP 1: IMPORT LIBRARIES

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

# ==========================================
# STEP 2: LOAD DATASET
# ==========================================

file_path = r"C:\Users\Ritesh\OneDrive\Desktop\Celebal Excellence Intership\Shopping Analysis\Data\archive\Combined_dataset.csv"

df = pd.read_csv(file_path)

# Standardize column names
df.columns = df.columns.str.strip().str.lower()

print("DATASET LOADED SUCCESSFULLY")

# ==========================================
# STEP 3: DATA EXPLORATION
# ==========================================

print("\n========== FIRST 5 ROWS ==========")
display(df.head())

print("\n========== LAST 5 ROWS ==========")
display(df.tail())

print("\n========== DATA SHAPE ==========")
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

print("\n========== COLUMN NAMES ==========")
print(df.columns.tolist())

print("\n========== DATA TYPES ==========")
print(df.dtypes)

print("\n========== DATASET INFO ==========")
df.info()

print("\n========== STATISTICAL SUMMARY ==========")
display(df.describe(include='all'))

# ==========================================
# STEP 4: MISSING VALUES
# ==========================================

print("\n========== MISSING VALUES ==========")
print(df.isnull().sum())

# Remove completely empty rows
df.dropna(how='all', inplace=True)

# ==========================================
# STEP 5: REMOVE DUPLICATES
# ==========================================

duplicate_count = df.duplicated().sum()

print("\nDuplicate Rows Found:", duplicate_count)

df.drop_duplicates(inplace=True)

print("Duplicate Rows Removed")

# ==========================================
# STEP 6: CONVERT NUMERIC COLUMNS
# ==========================================

numeric_columns = [
    'initial_price',
    'discount',
    'final_price',
    'rating',
    'ratings_count'
]

for col in numeric_columns:

    if col in df.columns:

        df[col] = pd.to_numeric(
            df[col].astype(str)
            .str.replace(r'[^0-9.\-]', '', regex=True),
            errors='coerce'
        )

# ==========================================
# STEP 7: HANDLE MISSING VALUES
# ==========================================

# Numeric columns -> Median
numeric_cols = df.select_dtypes(include='number').columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

# Text columns -> Unknown
text_cols = [
    col for col in df.columns
    if pd.api.types.is_object_dtype(df[col]) or pd.api.types.is_string_dtype(df[col])
]

df[text_cols] = df[text_cols].fillna("Unknown")

print("\nMissing Values Handled Successfully")

# ==========================================
# STEP 8: BASIC OPERATIONS
# ==========================================

# Filter Rows

if 'final_price' in df.columns:

    expensive_products = df[
        df['final_price'] > 1000
    ]

    print("\nProducts with Final Price > 1000")
    print("Count:", len(expensive_products))

    display(expensive_products.head())

# Select Columns

selected_cols = [
    col for col in
    [
        'title',
        'category',
        'initial_price',
        'final_price',
        'rating',
        'ratings_count'
    ]
    if col in df.columns
]

print("\n========== SELECTED COLUMNS ==========")

display(df[selected_cols].head())

# ==========================================
# STEP 9: FEATURE ENGINEERING
# ==========================================

# Price Difference

if (
    'initial_price' in df.columns and
    'final_price' in df.columns
):

    df['price_difference'] = (
        df['initial_price']
        - df['final_price']
    )

# Popularity Metric

if (
    'rating' in df.columns and
    'ratings_count' in df.columns
):

    df['popularity'] = (
        df['rating']
        * df['ratings_count']
    )

print("\nFeature Engineering Completed")

# ==========================================
# STEP 10: UNIVARIATE ANALYSIS
# ==========================================

print("\n========== UNIVARIATE ANALYSIS ==========")

if 'final_price' in df.columns:

    print("\nFinal Price Statistics")
    print(df['final_price'].describe())

if 'rating' in df.columns:

    print("\nRating Statistics")
    print(df['rating'].describe())

# ==========================================
# STEP 11: BIVARIATE ANALYSIS
# ==========================================

print("\n========== BIVARIATE ANALYSIS ==========")

if (
    'final_price' in df.columns and
    'rating' in df.columns
):

    correlation = df[
        ['final_price', 'rating']
    ].corr()

    print(correlation)

# ==========================================
# STEP 12: CATEGORY LEVEL ANALYSIS
# ==========================================

if (
    'category' in df.columns and
    'final_price' in df.columns
):

    category_analysis = (
        df.groupby('category')['final_price']
        .mean()
        .sort_values(ascending=False)
    )

    print("\n========== CATEGORY ANALYSIS ==========")

    print(category_analysis.head(10))

# ==========================================
# STEP 13: VISUALIZATION
# ==========================================

# Histogram

if 'final_price' in df.columns:

    plt.figure(figsize=(8,5))

    sns.histplot(
        df['final_price'],
        bins=30
    )

    plt.title("Final Price Distribution")

    plt.xlabel("Final Price")

    plt.ylabel("Frequency")

    plt.show()

# Boxplot

if 'final_price' in df.columns:

    plt.figure(figsize=(8,5))

    sns.boxplot(
        x=df['final_price']
    )

    plt.title("Final Price Boxplot")

    plt.show()

# Bar Chart

if (
    'category' in df.columns and
    'final_price' in df.columns
):

    category_analysis.head(10).plot(
        kind='bar',
        figsize=(10,5)
    )

    plt.title(
        "Top 10 Categories by Average Price"
    )

    plt.ylabel("Average Price")

    plt.show()

# Scatter Plot

if (
    'final_price' in df.columns and
    'rating' in df.columns
):

    plt.figure(figsize=(8,5))

    sns.scatterplot(
        x='final_price',
        y='rating',
        data=df
    )

    plt.title("Price vs Rating")

    plt.show()

# ==========================================
# STEP 14: INSIGHTS
# ==========================================

print("\n========== KEY INSIGHTS ==========")

if 'category' in df.columns:

    print("\nTop Categories:")

    print(
        df['category']
        .value_counts()
        .head(10)
    )

if 'final_price' in df.columns:

    print(
        f"\nAverage Final Price: "
        f"{df['final_price'].mean():.2f}"
    )

    print(
        f"Maximum Final Price: "
        f"{df['final_price'].max():.2f}"
    )

if 'rating' in df.columns:

    print(
        f"\nAverage Rating: "
        f"{df['rating'].mean():.2f}"
    )

# ==========================================
# STEP 15: SAVE CLEANED DATASET
# ==========================================

output_path = r"C:\Users\Ritesh\OneDrive\Desktop\Celebal Excellence Intership\Shopping Analysis\Data\Cleaned Dataset\Cleaned_Combined_dataset.csv"

df.to_csv(
    output_path,
    index=False
)

print("\nCleaned Dataset Saved Successfully")

print(output_path)

# ==========================================
# STEP 16: FINAL SUMMARY
# ==========================================

print("\n========== PROJECT SUMMARY ==========")

print(f"Total Rows: {df.shape[0]}")
print(f"Total Columns: {df.shape[1]}")

print(
    f"Remaining Missing Values: "
    f"{df.isnull().sum().sum()}"
)

print(
    f"Remaining Duplicate Rows: "
    f"{df.duplicated().sum()}"
)

print("\nProject Completed Successfully")
```

```python
# ==========================================
# SHOPPING DATASET ANALYSIS PROJECT
# ==========================================

# STEP 1: IMPORT LIBRARIES

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

# ==========================================
# STEP 2: LOAD DATASET
# ==========================================

file_path = r"C:\Users\Ritesh\OneDrive\Desktop\Celebal Excellence Intership\Shopping Analysis\Data\archive\Combined_dataset.csv"

df = pd.read_csv(file_path)

# Standardize column names
df.columns = df.columns.str.strip().str.lower()

print("DATASET LOADED SUCCESSFULLY")

# ==========================================
# STEP 3: DATA EXPLORATION
# ==========================================

print("\n========== FIRST 5 ROWS ==========")
display(df.head())

print("\n========== LAST 5 ROWS ==========")
display(df.tail())

print("\n========== DATA SHAPE ==========")
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

print("\n========== COLUMN NAMES ==========")
print(df.columns.tolist())

print("\n========== DATA TYPES ==========")
print(df.dtypes)

print("\n========== DATASET INFO ==========")
df.info()

print("\n========== STATISTICAL SUMMARY ==========")
display(df.describe(include='all'))

# ==========================================
# STEP 4: MISSING VALUES
# ==========================================

print("\n========== MISSING VALUES ==========")
print(df.isnull().sum())

# Remove completely empty rows
df.dropna(how='all', inplace=True)

# ==========================================
# STEP 5: REMOVE DUPLICATES
# ==========================================

duplicate_count = df.duplicated().sum()

print("\nDuplicate Rows Found:", duplicate_count)

df.drop_duplicates(inplace=True)

print("Duplicate Rows Removed")

# ==========================================
# STEP 6: CONVERT NUMERIC COLUMNS
# ==========================================

numeric_columns = [
    'initial_price',
    'discount',
    'final_price',
    'rating',
    'ratings_count'
]

for col in numeric_columns:

    if col in df.columns:

        df[col] = pd.to_numeric(
            df[col].astype(str)
            .str.replace(r'[^0-9.\-]', '', regex=True),
            errors='coerce'
        )

# ==========================================
# STEP 7: HANDLE MISSING VALUES
# ==========================================

# Numeric columns -> Median
numeric_cols = df.select_dtypes(include='number').columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

# Text columns -> Unknown
text_cols = [
    col for col in df.columns
    if pd.api.types.is_object_dtype(df[col]) or pd.api.types.is_string_dtype(df[col])
]

df[text_cols] = df[text_cols].fillna("Unknown")

print("\nMissing Values Handled Successfully")
# ==========================================
# STEP 8: BASIC OPERATIONS
# ==========================================

# Filter Rows

if 'final_price' in df.columns:

    expensive_products = df[
        df['final_price'] > 1000
    ]

    print("\nProducts with Final Price > 1000")
    print("Count:", len(expensive_products))

    display(expensive_products.head())

# Select Columns

selected_cols = [
    col for col in
    [
        'title',
        'category',
        'initial_price',
        'final_price',
        'rating',
        'ratings_count'
    ]
    if col in df.columns
]

print("\n========== SELECTED COLUMNS ==========")

display(df[selected_cols].head())

# ==========================================
# STEP 9: FEATURE ENGINEERING
# ==========================================

# Price Difference

if (
    'initial_price' in df.columns and
    'final_price' in df.columns
):

    df['price_difference'] = (
        df['initial_price']
        - df['final_price']
    )

# Popularity Metric

if (
    'rating' in df.columns and
    'ratings_count' in df.columns
):

    df['popularity'] = (
        df['rating']
        * df['ratings_count']
    )

# ==========================================
# ✅ ADDED: total_amount = final_price × quantity
# ==========================================

if (
    'final_price' in df.columns and
    'quantity' in df.columns
):

    df['total_amount'] = (
        df['final_price']
        * df['quantity']
    )

    print("\ntotal_amount column created (final_price x quantity)")

else:

    # Fallback if dataset has no quantity column
    df['total_amount'] = df['final_price'] * 1

    print("\nNote: No 'quantity' column found.")
    print("total_amount set equal to final_price as fallback.")

print("\nFeature Engineering Completed")
print("New Columns Added: price_difference, popularity, total_amount")

# ==========================================
# STEP 10: UNIVARIATE ANALYSIS
# ==========================================

print("\n========== UNIVARIATE ANALYSIS ==========")

if 'final_price' in df.columns:

    print("\nFinal Price Statistics")
    print(df['final_price'].describe())

if 'rating' in df.columns:

    print("\nRating Statistics")
    print(df['rating'].describe())

# ==========================================
# STEP 11: BIVARIATE ANALYSIS
# ==========================================

print("\n========== BIVARIATE ANALYSIS ==========")

if (
    'final_price' in df.columns and
    'rating' in df.columns
):

    correlation = df[
        ['final_price', 'rating']
    ].corr()

    print(correlation)

# ==========================================
# STEP 12: CATEGORY LEVEL ANALYSIS
# ==========================================

if (
    'category' in df.columns and
    'final_price' in df.columns
):

    category_analysis = (
        df.groupby('category')['final_price']
        .mean()
        .sort_values(ascending=False)
    )

    print("\n========== CATEGORY ANALYSIS ==========")

    print(category_analysis.head(10))

# ==========================================
# STEP 13: VISUALIZATION
# ==========================================

# Histogram

if 'final_price' in df.columns:

    plt.figure(figsize=(8, 5))

    sns.histplot(
        df['final_price'],
        bins=30
    )

    plt.title("Final Price Distribution")

    plt.xlabel("Final Price")

    plt.ylabel("Frequency")

    plt.tight_layout()

    plt.show()

# Boxplot

if 'final_price' in df.columns:

    plt.figure(figsize=(8, 5))

    sns.boxplot(
        x=df['final_price']
    )

    plt.title("Final Price Boxplot")

    plt.tight_layout()

    plt.show()

# Bar Chart

if (
    'category' in df.columns and
    'final_price' in df.columns
):

    category_analysis.head(10).plot(
        kind='bar',
        figsize=(10, 5)
    )

    plt.title(
        "Top 10 Categories by Average Price"
    )

    plt.ylabel("Average Price")

    plt.tight_layout()

    plt.show()

# Scatter Plot

if (
    'final_price' in df.columns and
    'rating' in df.columns
):

    plt.figure(figsize=(8, 5))

    sns.scatterplot(
        x='final_price',
        y='rating',
        data=df
    )

    plt.title("Price vs Rating")

    plt.tight_layout()

    plt.show()

# ==========================================
# STEP 14: INSIGHTS
# ==========================================

print("\n========== KEY INSIGHTS ==========")

if 'category' in df.columns:

    print("\nTop Categories by Product Count:")

    print(
        df['category']
        .value_counts()
        .head(10)
    )

if 'final_price' in df.columns:

    print(
        f"\nAverage Final Price: "
        f"{df['final_price'].mean():.2f}"
    )

    print(
        f"Maximum Final Price: "
        f"{df['final_price'].max():.2f}"
    )

if 'rating' in df.columns:

    print(
        f"\nAverage Rating: "
        f"{df['rating'].mean():.2f}"
    )

if 'total_amount' in df.columns:

    print(
        f"\nTotal Revenue (all products): "
        f"{df['total_amount'].sum():.2f}"
    )

    print(
        f"Average Order Value: "
        f"{df['total_amount'].mean():.2f}"
    )

if 'price_difference' in df.columns:

    print(
        f"\nAverage Price Saving per Product: "
        f"{df['price_difference'].mean():.2f}"
    )

# ==========================================
# ✅ ADDED: BUSINESS IMPLICATIONS
# ==========================================

print("\n========== BUSINESS IMPLICATIONS ==========")

print("""
1. PRICING STRATEGY:
   High-priced categories (e.g. Electronics) should be
   prioritized for seasonal discount campaigns to drive
   volume without heavy margin loss.

2. PRODUCT RECOMMENDATIONS:
   Products with high popularity scores (rating x ratings_count)
   are strong candidates for homepage featuring and
   search ranking boosts.

3. DISCOUNT MANAGEMENT:
   Large price differences (initial vs final price) indicate
   aggressive discounting. Monitor these SKUs for profit
   margin erosion over time.

4. QUALITY CONTROL:
   Categories or products with consistently low ratings
   need supplier review, quality checks, or should be
   delisted to protect brand trust.

5. INVENTORY PLANNING:
   total_amount (price x quantity) reveals which products
   drive the most revenue. High total_amount products
   should maintain strong stock availability.

6. PREMIUM SEGMENT TARGETING:
   Categories with few products but high average price
   represent premium segments. These need targeted
   marketing rather than broad discount offers.
""")

# ==========================================
# STEP 15: SAVE CLEANED DATASET
# ==========================================

output_path = r"C:\Users\Ritesh\OneDrive\Desktop\Celebal Excellence Intership\Shopping Analysis\Data\Cleaned Dataset\Cleaned_Combined_dataset.csv"

df.to_csv(
    output_path,
    index=False
)

print("\nCleaned Dataset Saved Successfully")

print(output_path)

# ==========================================
# STEP 16: FINAL SUMMARY
# ==========================================

print("\n========== PROJECT SUMMARY ==========")

print(f"Total Rows        : {df.shape[0]}")
print(f"Total Columns     : {df.shape[1]}")

print(
    f"Missing Values    : "
    f"{df.isnull().sum().sum()}"
)

print(
    f"Duplicate Rows    : "
    f"{df.duplicated().sum()}"
)

print(
    f"New Columns Added : "
    f"price_difference, popularity, total_amount"
)

print("\nProject Completed Successfully")
```
