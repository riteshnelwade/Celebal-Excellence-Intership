```python
import pandas as pd
import numpy as np
from IPython.display import display

# ==========================================
# 1. LOAD DATASET
# ==========================================

file_path = r"C:\Users\Ritesh\OneDrive\Desktop\Celebal Excellence Intership\Shopping Analysis\Data\archive\Combined_dataset.csv"
df = pd.read_csv(file_path)

# Standardize column names
df.columns = df.columns.str.strip().str.lower()

print("DATASET LOADED SUCCESSFULLY")
print("\nColumns in dataset:")
print(df.columns.tolist())

# ==========================================
# 2. DATA EXPLORATION
# ==========================================

print("\n========== FIRST 5 ROWS ==========")
display(df.head())

print("\n========== LAST 5 ROWS ==========")
display(df.tail())

print("\n========== DATA SHAPE ==========")
print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")

print("\n========== DATA TYPES ==========")
print(df.dtypes)

print("\n========== DATASET INFO ==========")
df.info()

print("\n========== STATISTICAL SUMMARY ==========")
display(df.describe(include='all'))

# ==========================================
# 3. MISSING VALUES
# ==========================================

print("\n========== MISSING VALUES ==========")
print(df.isnull().sum())

# Remove completely empty rows
df = df.dropna(how='all')

# ==========================================
# 4. REMOVE DUPLICATES
# ==========================================

duplicate_count = df.duplicated().sum()
print(f"\nDuplicate Rows Found: {duplicate_count}")

if duplicate_count > 0:
    df = df.drop_duplicates()
    print("Duplicates Removed Successfully")
else:
    print("No Duplicates Found")

# ==========================================
# 5. CLEAN / CONVERT NUMERIC COLUMNS
# ==========================================

numeric_columns = ['initial_price', 'discount', 'final_price', 'rating', 'ratings_count']

for col in numeric_columns:
    if col in df.columns:
        df[col] = pd.to_numeric(
            df[col].astype(str).str.replace(r'[^0-9.\-]', '', regex=True),
            errors='coerce'
        )

# ==========================================
# 6. HANDLE MISSING VALUES
# ==========================================

# Fill numeric columns with median
num_cols = df.select_dtypes(include='number').columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Fill text columns with "Unknown"
text_cols = df.select_dtypes(include=['object', 'string']).columns
df[text_cols] = df[text_cols].fillna("Unknown")

print("\nMissing Values Handled Successfully")

# ==========================================
# 7. BASIC OPERATIONS
# ==========================================

# Filter rows: products with final_price > 1000
if 'final_price' in df.columns:
    expensive_products = df[df['final_price'] > 1000]
    print("\n========== FILTERED ROWS ==========")
    print(f"Products with Final Price > 1000: {len(expensive_products)}")
    display(expensive_products.head())

# Select columns
selected_cols = [col for col in ['title', 'category', 'initial_price', 'final_price', 'rating', 'ratings_count'] if col in df.columns]

if selected_cols:
    print("\n========== SELECTED COLUMNS ==========")
    display(df[selected_cols].head())

# ==========================================
# 8. DERIVED COLUMN
# ==========================================

# Assignment says total_amount = price * quantity
# This dataset does not have price and quantity columns.
# So we create a meaningful derived column instead:
# saved_amount = initial_price - final_price

if 'initial_price' in df.columns and 'final_price' in df.columns:
    df['saved_amount'] = df['initial_price'] - df['final_price']
    print("\n========== DERIVED COLUMN ==========")
    print("saved_amount = initial_price - final_price")

# ==========================================
# 9. BASIC INSIGHTS
# ==========================================

print("\n========== INSIGHTS ==========")

if 'category' in df.columns:
    print("\nTop 10 Categories:")
    print(df['category'].value_counts().head(10))

if 'final_price' in df.columns:
    print(f"\nAverage Final Price: {df['final_price'].mean():.2f}")
    print(f"Maximum Final Price: {df['final_price'].max():.2f}")

if 'rating' in df.columns:
    print(f"\nAverage Rating: {df['rating'].mean():.2f}")

# ==========================================
# 10. SAVE CLEANED DATASET
# ==========================================

output_path = r"C:\Users\Ritesh\OneDrive\Desktop\Celebal Excellence Intership\Shopping Analysis\Data\Cleaned Dataset\Cleaned_Combined_dataset_v2.csv"
df.to_csv(output_path, index=False)

print("\nCleaned Dataset Saved Successfully")
print(output_path)

# ==========================================
# 11. SUMMARY
# ==========================================

print("\n========== DATA CLEANING SUMMARY ==========")
print(f"Total Rows: {df.shape[0]}")
print(f"Total Columns: {df.shape[1]}")
print(f"Remaining Missing Values: {df.isnull().sum().sum()}")
print(f"Remaining Duplicate Rows: {df.duplicated().sum()}")
print("\nProject Completed Successfully")
```
